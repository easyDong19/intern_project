```python
import pandas as pd
import numpy as np
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
AAPL_data = pd.read_csv('data/AAPL.csv')
GSPC_data = pd.read_csv('data/GSPC.csv')
AAPL_data.head()
```

```python
AAPL_data['log_Return'] = np.log(AAPL_data['Adj Close']).diff(1)
GSPC_data['log_Return'] = np.log(GSPC_data['Adj Close']).diff(1)

GSPC_data.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,10))
plt.plot(AAPL_data['log_Return'], label='AAPL', color='red')

plt.yticks(np.arange(-0.8,1,0.2))
plt.xticks(np.arange(0,4000,500))

plt.xlim(0,3500)

plt.grid(True)

plt.legend()
plt.xlabel('Time(day)')
plt.ylabel('Daily returns')
plt.savefig('그래프1.png')

plt.show()
```

```python
from scipy.stats import norm

plt.figure(figsize=(15,10))


plt.xticks(np.arange(-0.10,0.20,0.05))
plt.xlim(-0.10,0.15)
plt.ylim(0,50)

plt.hist(GSPC_data['log_Return'].dropna(), bins=50, density=True,edgecolor='black', linewidth=1.5)
mean = GSPC_data['log_Return'].mean()
std = GSPC_data['log_Return'].std()

x = np.arange(-1, 1,0.001)
plt.plot(x, norm.pdf(x, loc = mean, scale=std), linestyle="--", color="red")

plt.legend()
plt.xlabel('Returns')
plt.ylabel('Normal and Estimated Density')
plt.savefig('그래프3.png')


plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

## ADF
def ADF(data):

    result = adfuller(GSPC_data['log_Return'].dropna(), autolag="AIC")

    print("---- Adfuller ----")
    print('ADF Statistic: %f' % result[0])
    print('p-value: %1.10f' % result[1])
    print('Lag: %d' % result[2])
    print('observation: %d' % result[3])
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t%s: %.3f' % (key, value))

from statsmodels.tsa.stattools import kpss

def kpss_test(x, h0_type = 'c') :
    indices = ['Test Statistics', 'p-value', '# of Lags Used']
    kpss_test = kpss(x, regression = h0_type)
    results = pd.Series(kpss_test[0:3], index = indices)

    for key, value in kpss_test[3].items() :
        results[f'Critical Value ({key})'] = value

    return results
# ADF(GSPC_data['log_Return'].dropna())
kpss_test(GSPC_data['log_Return'].dropna())
```
